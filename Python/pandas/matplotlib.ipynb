```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[8,3,5,6]
plt.plot(x,y,marker='o',linestyle='--',color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
x1=[1,2,3,4]
x2=[10,2,33,4]
y1=[10,20,30,40]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x1,x2,color='red')

plt.subplot(2,2,2)
plt.plot(x1,y1,color='green')

plt.subplot(2,2,3)
plt.plot(x2,y1,color='green')

plt.subplot(2,2,4)
plt.plot(x1,y1,color='blue')
# plt.subplot(2,1,5) it is telling maximum number of plot in a graph
# plt.plot(x1,y1,color='red') not possible 
```
